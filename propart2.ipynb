```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
mall_customers = pd.read_csv('Mall_Customers.csv')
print(mall_customers.head())
print(mall_customers.shape)
print(mall_customers.describe())
print(mall_customers.info())
```

```python
missing_values = mall_customers.isnull().sum()
missing_values[missing_values > 0]
```

```python
gender_counts = mall_customers['Gender'].value_counts()
print(gender_counts)
```

```python
plt.figure(figsize=(6,6))
plt.pie(gender_counts, labels=['Normal (0)', 'Fraud (1)'], autopct='%1.2f%%', colors=['skyblue', 'red'])
plt.title('Distribution of Transaction Classes')
```

```python
Mall = mall_customers.drop(columns=['Gender'])
scaler = StandardScaler()
Mall_scaled = scaler.fit_transform(Mall)
Mall_scaled
```

```python
Mall_scaled_df = pd.DataFrame(Mall_scaled, columns=Mall.columns)
Mall_scaled_df.hist(bins=50, figsize=(20,15), color='pink')
plt.suptitle("Feature Distributions AFTER Scaling")
plt.tight_layout()
```
